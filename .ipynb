```python
data = []
nama = ""
kategori = ""
harga = ""

def inputData():
    nama = input("Masukkan nama barang: ")
    kategori = input("Masukkan kategori barang: ")
    harga = input("Masukkan harga barang: Rp ")
    data.append([nama,kategori,harga])
    print(f"{nama} berhasil ditambahkan.")

def menampilkanData():
    if not data:
        print("Data belum tersedia, silahkan pilih menu 1 untuk isi data.")
    else:
        print("No|Nama barang|Kategori barang|Harga")
        i = [nama,kategori,harga]
        n = 0
        
        for i in range(len(data)):
            n += 1
            print (n,"|",data[i][0],"|",data[i][1],"|",data[i][2])

def runningProgram():
    pilihanMenu = 0
    while True:

        print("Menu:")
        print("1. Input data")
        print("2. Tampilkan data")
        print("pilih menu atau keluar: ")
    
        pilihanMenu = input("Pilih menu: ")

        if pilihanMenu == "1":
            inputData()
        elif pilihanMenu == "2":
            menampilkanData()
        elif pilihanMenu == "exit":
            print("Terimakasih, sampai jumpa!")
            break

runningProgram()

    
```

```python
# List untuk menyimpan data inventory barang
inventory = []

# Fungsi untuk menambah barang ke dalam inventory
def tambah_barang():
    nama = input("Masukkan nama barang: ")
    kategori = input("Masukkan kategori barang: ")
    try:
        harga = float(input("Masukkan harga barang: Rp "))
        # Tambahkan data barang sebagai list [nama, kategori, harga] ke inventory
        inventory.append([nama, kategori, harga])
        print(f"{nama} berhasil ditambahkan ke inventory.\n")
    except ValueError:
        print("Harga harus berupa angka. Coba lagi.\n")

# Fungsi untuk menampilkan seluruh inventory
def tampilkan_inventory():
    if not inventory:
        print("Data belum ada.\n")
    else:
        print("\n=== Inventory Stock Barang ===")
        print("No.\t| Nama Barang     | Kategori          | Harga")
        print("----|----------------|-------------------|---------")
        for i, item in enumerate(inventory, start=1):
            print(f"{i}\t| {item[0]:<15} | {item[1]:<17} | Rp {item[2]:,.2f}")
        print()  # Print new line setelah tampilan inventory

# Fungsi utama untuk menjalankan sistem
def main():
    while True:
        print("Menu:")
        print("1. Tambah Barang")
        print("2. Tampilkan Inventory")
        print("Ketik 'exit' untuk keluar.")
        
        pilihan = input("Pilih menu: ")
        
        if pilihan == "1":
            tambah_barang()
        elif pilihan == "2":
            tampilkan_inventory()
        elif pilihan.lower() == "exit":
            print("Program dihentikan.")
            break
        else:
            print("Pilihan tidak valid, silakan coba lagi.\n")

# Menjalankan program
main()
```
